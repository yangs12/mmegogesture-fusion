```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import metrics
import glob
import os
```

```python
def plot_confusion(y_ls, y_pred_ls, labels, name):
    """Save the confusion matrix and des information
    """
    font = {'size': 10}
    plt.rc('font', **font)
    plt.rcParams['figure.figsize'] = [10, 8]
    cm_display =  metrics.ConfusionMatrixDisplay.from_predictions(
        y_ls, 
        y_pred_ls, 
        values_format = '.3f',
        display_labels= labels,
        normalize='true', 
        include_values=True, 
        cmap=plt.cm.Blues, 
        colorbar=False
        )
    plt.xticks(rotation=30)
    plt.title(name) 
    # plt.savefig(path_des+'fig/cm-'+name+'.png')
    # plt.close()
    plt.show()
  
```

```python

# Define path to CSV files
des_files_path = '/home/jchang13/fusion/mmegogesture-fusion/vis_outputs/csv/concat'

# Use glob to get all CSV files in the directory
csv_files = glob.glob(os.path.join(des_files_path, '*.csv'))

# Read and collect all DataFrames
des_df_ls = [pd.read_csv(f) for f in csv_files]

# Concatenate into one DataFrame
des_df = pd.concat(des_df_ls, axis=0, ignore_index=True)

# Optional: print or check shape
print(f"Loaded and concatenated {len(csv_files)} CSV files.")

y_ls = des_df['test_real']
y_pred_ls = des_df['test_prediction']

print(len(y_ls))
print(len(y_pred_ls))
labels = [ "Click", "Snap", "ZIn", "Knock", "SwpL", "SwpR", "T_ZIn", "T_ZOut", "Stop", "OK", "TmbUp", "Photo" ]

plot_confusion(y_ls, y_pred_ls, labels, 'Average confusion matrix of Concat Method')
accuracy = metrics.accuracy_score(y_ls, y_pred_ls)
print(f'Average accuracy: {accuracy:.3f}')

```
